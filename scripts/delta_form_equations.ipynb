# Script used to derive the delta form equations

```python
import sympy as sp
```

```python
# 1. Definindo os coeficientes congelados no tempo n (Tratados como constantes)
alpha, omega = sp.symbols('alpha omega ')

# 2. Definindo a variável dependente (Delta r) e suas derivadas espaciais literais
dr = sp.Symbol('Delta_r')

# Operador resídio
L_r = sp.Symbol('L_r')

# Coefs.
A = sp.Symbol('A')
B = sp.Symbol('B')
C = sp.Symbol('C')

# Definindo os pontos de malha para x e y
dr_ij = sp.Symbol('Delta_r_ij')
dr_ip1j = sp.Symbol('Delta_r_i+1,_j')
dr_im1j = sp.Symbol('Delta_r_i-1,_j')
dr_ijp1 = sp.Symbol('Delta_r_i,_j+1')
dr_ijm1 = sp.Symbol('Delta_r_i,_j-1')
dr_im1jm1 = sp.Symbol('Delta_r_i-1,_j-1')
dr_ip1jp1 = sp.Symbol('Delta_r_i+1,_j+1')
dr_ip1jm1 = sp.Symbol('Delta_r_i+1,_j-1')
dr_im1jp1 = sp.Symbol('Delta_r_i-1,_j+1')
```

```python
delta_eq = - 1/2 * (alpha * dr_ij - A * (dr_ip1j - 2 * dr_ij + dr_im1j)) * (alpha*dr_ij - C * (dr_ijp1 - 2 * dr_ij + dr_ijm1)) + omega * L_r
delta_eq
```

```python
delta_eq_exp = sp.expand(delta_eq)
delta_eq_exp
```

```python
delta_eq_exp.coeff(dr_ij)
```
